```python
Ideal_Weather= city_data_df.loc[(city_data_df['Max Temp'] >=280 ) & (city_data_df['Max Temp'] <= 380) & 
(city_data_df['Wind'] <= 2) & (city_data_df['Humidity'] <= 50) &(city_data_df['Cloudiness']<=50)]
#Ideal_Weather.reset_index(drop=True, inplace=True)
Ideal_Weather
```

```python
hotel_data=Ideal_Weather[["City","Country","Lat","Lng","Humidity",]].copy()
hotel_data["Hotel Name"] = ''
hotel_data
```

```python
url="https://api.geoapify.com/v2/places/json"

params = {
    "radius": 50000,
    "rankby": "prominence",
    "types": "lodging",
    "keyword": "hotel",
    "key": geoapify_key,
}

for index, row in hotel_data.iterrows():
    lat = row['Lat']
    lng = row['Lng']
    city = row['City']
    params["location"] = f'{lat},{lng}'
        
    name_address = requests.get(url, params=params).json()
    
   
    
try:
        hotel_data.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
        
except(KeyError, IndexError):
        hotel_data.loc[index, "Hotel Name"] = "No hotel found"
        print(f"{hotel_data.loc[index, 'City']} - nearest hotel: {hotel_data.loc[index, 'Hotel Name']}")
        
print("------------")

hotel_data
```
